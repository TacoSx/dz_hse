```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import time
import pandas as pd
```

# Фрагмент кода запускался, и останавливался вручную в бесконечном цикле, после остановки обрабатывался Ексепшн .sleep() записывалось время остановки, а дальше вниз по коду добавлялось все в файл без дубликатов и выводилась нужная статистика.

```python
alive_sites = []
```

```python
now = datetime.now()
while(True):
    url = "https://openphish.com/"
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    
    if page.status_code == 200:
        table = soup.find("table", class_ = 'pure-table pure-table-striped')
        iternal_table = table.find('tbody')   
    
    now2 = datetime.now()
    currect_time = now2.strftime('%m/%d/%Y %Y:%M:%S')
    date = currect_time.split(" ")[0]

    for tr in iternal_table.find_all('tr'):
        url_tr = ""
        target = ""
        time_tr = ""
        row = []
        for td in tr.find_all('td'):
            row.append(td.text.strip())
        print(row)
        if row:
            url_tr = row[0]
            target = row[1]
            time_tr = date + " " +row[2]

            datetime_object = datetime.strptime(time_tr,'%m/%d/%Y %H:%M:%S')
            alive_sites.append(row)
    df = pd.DataFrame(alive_sites)
    try:
        time.sleep(600)
    except KeyboardInterrupt:
        now2 = datetime.now()
        break;
```

# В файле от 0 до 29(первые 30) это самые первые ссылки, последующие добавлялись не в начало списка,а в конец. Поэтому последующие новые добавления идут после 29 индекса. Но это не слишком,то и важно.
